```python
import torch
```

```python
banch_n = 100
hidden_layer = 100
input_data = 1000
output_data = 10
```

```python
x = torch.randn(banch_n, input_data)  #(100*1000)
y = torch.randn(banch_n, output_data)  #(100*10)

w1 = torch.randn(input_data, hidden_layer)  #(1000*100)
w2 = torch.randn(hidden_layer, output_data)  #(100*10)
```

```python
epoch_n = 50000
learning_rate = 1e-6
```

```python
for epoch in range(epoch_n):
    h1 = x.mm(w1)  #(100*1000)*(1000*100)
#     print(h1.size())
    h1 = h1.clamp(min = 0)
    y_pred = h1.mm(w2)  #(100*100)*(100*10)
    
    loss = (y_pred - y).pow(2).sum()
    if epoch%100 == 0 :
        print('Epoch:{}, Loss:{:.4f}'.format(epoch,loss))
    grad_y_pred = 2 * (y_pred - y)
    grad_w2 = h1.t().mm(grad_y_pred)
    
    grad_h = grad_y_pred.clone()
    grad_h = grad_h.mm(w2.t())
    grad_h.clamp(min=0)
    grad_w1 = x.t().mm(grad_h)
    
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
```
